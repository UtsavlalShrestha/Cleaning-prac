```python
#renaming some columns
df = df.rename(columns = {'Mat':'matches','NO': 'not_outs', 'HS': 'highest_inn_score', 'BF':'balls_faced', 'SR': 'strike_rate'}) 
df.head()
```

```python
#checking for null values
df.isnull().any()
```

```python
df[df['balls_faced'].isna()==1]
```

```python
df['balls_faced'] = df['balls_faced'].fillna(0)
```

```python
df['strike_rate'] = df['strike_rate'].fillna(0)
```

```python
#dropping duplicates
```

```python
df= df.drop_duplicates()
```

```python
df[df['Player'].isin(['GA Headley (WI)','GS Sobers (WI)', 'JB Hobbs (ENG)', 'V Kohli (IND)'])]
```

```python
#splitting span as starting and ending
```

```python
df['debut'] = df['Span'].str.split('-').str.get(0)
```

```python
df['retired'] = df['Span'].str.split('-').str.get(1)
```

```python
df.head()
```

```python
df = df.drop(['Span'], axis = 1)
```

```python
df.head()
```

```python
#splitting country from player and removing country code from under the player name
```

```python
df['country'] = df['Player'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df.head()
```

```python
df['Player'] = df['Player'].str.split('(').str.get(0)
```

```python
df.head()
```

```python
#replacing data types
```

```python
df.dtypes
```

```python
df['highest_inn_score'] = df['highest_inn_score'].str.split('*').str.get(0)
```

```python
df['highest_inn_score'] = df['highest_inn_score'].astype(int)
```

```python
df.dtypes
```

```python
df = df.astype({'debut': 'int', 'retired': 'int'})
```

```python
#got error thus clean that +, also had some null values
df['balls_faced'] = df['balls_faced'].str.split('+').str.get(0)
```

```python
df['balls_faced'] = df['balls_faced'].fillna(0).astype(int)
```

```python
df = df.astype({'balls_faced' : 'int', 'strike_rate':'int'})
```

```python
df.dtypes
```

```python
#adding career length column
```

```python
df['career_span'] = df['retired']-df['debut']
```

```python
df.head()
```

```python
#what is lowest avergae career span of cricketers based on countries
df.groupby('country')['career_span'].mean().sort_values(ascending=True).head()
```

```python
#what is average batting strike rate those who player over 10 years
df[df['career_span'] >10]['strike_rate'].mean()
```

```python
#number of players before 1960
len(df.query('retired < 1960'))
```

```python
#max highest innings score by country
df.groupby('country')['highest_inn_score'].max().to_frame('highest_in_the_country')
```
